# PART 1: Import Libraries

```

import matplotlib.pyplot as plt
import numpy as np
```

# PART 2: Create Sample Dataset

```

students = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen", "Ian", "Jane"]
math_scores = [65, 80, 45, 90, 75, 55, 88, 92, 60, 73]
english_scores = [70, 85, 50, 88, 80, 58, 90, 95, 62, 77]
science_scores = [72, 78, 48, 92, 79, 60, 85, 98, 64, 74]
```


# PART 3: Histogram

A histogram shows the distribution of numerical data.


```

plt.figure()
plt.hist(math_scores, bins=5)
plt.title("Distribution of Math Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```


# PART 4: Scatter Plot

A scatter plot shows the relationship between two variables.


```

plt.figure()
plt.scatter(math_scores, english_scores)
plt.title("Math vs English Scores")
plt.xlabel("Math Scores")
plt.ylabel("English Scores")
plt.show()
```


# PART 5: Multiple Line Chart

Compare trends of multiple subjects.


```

plt.figure()
plt.plot(students, math_scores)
plt.plot(students, english_scores)
plt.title("Math and English Scores Comparison")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```


# PART 6: Subplots

Display multiple charts in one figure.


```

plt.figure()

plt.subplot(1, 2, 1)
plt.hist(science_scores, bins=5)
plt.title("Science Histogram")

plt.subplot(1, 2, 2)
plt.scatter(math_scores, science_scores)
plt.title("Math vs Science")

plt.tight_layout()
plt.show()
```
